```python
#pseudonymize username
import hashlib
def consistent_hash(name, length=8):
    # Use SHA256 for consistent hashing
    if pd.isna(name):  
        return "N/A"
    if not isinstance(name, str):
        name = str(name)
    hash_object = hashlib.sha256(name.encode())
    return hash_object.hexdigest()[:length]
to_match = ["mentioning_users/0",
    "mentioning_users/1",
    "mentioning_users/2",
    "mentioning_users/3",
    "mentioning_users/4",
    "mentioning_users/5",
    "mentioning_users/6",
    "mentioning_users/7",
    "mentioning_users/8",
    "mentioning_users/9",
    "mentioning_users/10",
    "mentioning_users/11",
    "mentioning_users/12",
    "mentioning_users/13",
    "mentioning_users/14",
    "mentioning_users/15",
                             "replying_to/0",
    "replying_to/1",
    "replying_to/2",
    "replying_to/3",
    "replying_to/4",
    "replying_to/5",
    "replying_to/6",
    "replying_to/7",
    "replying_to/8",
    "replying_to/9",
    "replying_to/10",
    "replying_to/11",
    "replying_to/12",
    "replying_to/13",
    "replying_to/14",
    "replying_to/15",
    "replying_to/16",
    "replying_to/17",
    "replying_to/18",
    "replying_to/19",
    "replying_to/20",
    "replying_to/21",
    "replying_to/22",
    "replying_to/23",
    "replying_to/24",
    "replying_to/25",
    "replying_to/26",
    "replying_to/27",
    "replying_to/28",
    "replying_to/29",
    "replying_to/30",
    "replying_to/31",
    "replying_to/32",
    "replying_to/33",
    "replying_to/34",
    "replying_to/35",
    "replying_to/36",
    "replying_to/37",
    "replying_to/38",
    "replying_to/39",
    "replying_to/40",
    "replying_to/41",
    "replying_to/42",
    "replying_to/43",
    "replying_to/44",
    "replying_to/45",
    "replying_to/46",
    "replying_to/47", "user_name"]
unique_names = pd.unique(df[to_match].values.ravel())
name_to_mask = {name: consistent_hash(name) for name in unique_names}
for column in to_match:
    df[column] = df[column].map(name_to_mask)
```
